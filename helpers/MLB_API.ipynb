```python
import statsapi
import json
import re
from pybaseball import team_batting
```

```python
sched = statsapi.schedule(start_date='05/24/2024',end_date='05/24/2024')

def getKStats(raw_data, name):
    lines = [line.strip() for line in raw_data.split('\n') if line.strip()]
    relevant_lines = lines[2:]
    data = {}
    for line in relevant_lines:
        key, value = line.split(':')
        data[key.strip()] = value.strip()
    print(data)    
    specific_fields = {
        "Name": name,
        "GP": data.get("gamesPlayed"),
        "K": data.get("strikeOuts"),
        "AVG": data.get("avg"),
        "S%": data.get("strikePercentage"),
        "P/I": data.get("pitchesPerInning"),
        "K/9": data.get("strikeoutsPer9Inn")
    }
    json_data = json.dumps(specific_fields, indent=4)
    return json_data
    
def getPitcherStats(name):
    try:
        player = statsapi.lookup_player(name)
        id = player[0]['id']
        stats = statsapi.player_stats(id, group="[pitching]", type="season")
        try:
            short = getKStats(stats, name)
            return short
        except:
            return "ERROR: cannot shorten pitcher stats \n"
    except:
        return "ERROR: cannot find pitcher stats \n"
    
for game in sched:
    away_team = game['away_name']
    home_team = game['home_name']
    away_pitcher = game['away_probable_pitcher']
    home_pitcher = game['home_probable_pitcher']
    print(f"Away Team: {away_team}")
    print(getPitcherStats(away_pitcher))
    print(f"Home Team: {home_team}")
    print(getPitcherStats(home_pitcher))
    print("----------------------------------------------------------------------------------------")    
```
